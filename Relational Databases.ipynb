**For this relational database, let's first establish some assumptions:**

+ 'Purchase_id' is a unique identifier for each transaction. 


+ For each purchase that is made, we have identified the customer who makes the purchase, i.e. the customer_id field in the 'Sales' table cannot contain null values. 


+ The 'sales' table reflects the transactions by 'product_id', that is to say, if a customer buys two products with different id, we will compute two entries, one for each product, so we will always see the breakdown of each product that each customer has bought. 


+ Product_id' is a unique identifier for the product type but there can be two products that are equal and in this case they will have the same 'product_id' number. 


+ employees work in only one store, in this database there are no employees working in two stores at the same time.

Some considerations to take into account: 
    
+ **Primay Key:** The primary key in each table has been underlined and writen in red with the letters PK next to it. 
+ **Foreign Key** : The foreign key in each table is represented by the acronym FK in parentheses next to the name of the column that represents the key. 
+ **Relationships** : We have represented the relationships with arrows that indicate the relationship between the foreign key of one table and the primary key of the other.  We have put a single arrow on the side where the relationship is 1 and a double arrow when the relationship is multiple. 

**Explanation of relationships:**

+ 'Sales' has a 1:N relationship with Customers: a transaction can be performed by only one customer but the same customer can perform many transactions. 


+ Sales' has a N:1 relationship with stores: One transaction is performed in one store, but one store performs many transactions. 


+ Sales' and 'Employees' have an N:1 relationship: Each transaction is performed by a single employee, but that employee may perform many sales. 


+ Sales' and 'Products': N:1 relationship: each sale corresponds to a 'product_id', although the same 'product_id' can appear several times in the 'sales' table. 


+ Stores' and 'Employees' :their ratio is 1:N, an employee can work only in one store but several employees work in the same store. 


+ Products' and 'Product Suppliers': the relationship between the two tables is N:1, a product_id has only one manufacturer/supplier but the same manufacturer/supplier can provide several products with different product_id.

The only difference with the relational model of the previous exercise is that to define the relationships in this case **we have used a rhombus and changed the arrows for the standard symbols that define the cardinality of the relationships.** 

**Additional explanations:**
    
+ Entities are represented inside a rectangle, attributes inside a circle and relationships are described inside a diamond. 

+ When the circle of the attribute is in dashed line it means that the is a derived attribute, it is not necessary to store it in the database, it can be calculated from another attribute. 

+ When the circle of the attribute is double it means that it can contain more than one value. 

+ The underlined attribute refers to the primary key (value must be unique within that table). 

+ If the attribute contains (O) it means that we can leave it blank, it is optional. 

+ If the attribute name is inside parenthesis, it means that it can contain several fields (e.g. product description). 

+ Inside the rhombus we have tried to briefly describe the relationship by means of the action that links the two entities.
